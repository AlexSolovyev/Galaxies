### Посмотреть на данные

```python
from astropy.io import fits
```

```python
# открыть файлы с скрасными смещениями

hdul = fits.open('http://gal-03.sai.msu.ru/~vtoptun/redshift/rcsed_v2.fits')

hdul.info()
```

```python
hdr0 = hdul[0].header
hdr0
```

```python
hdr1 = hdul[1].header
hdr1
```

```python
# признаки
cnt = 1
features = []
print('Признаки:')
for key in hdr1.keys():
    if key[:2] == 'TT':
        print('    ', cnt, '-', hdr1[key])
        features.append(hdr1[key])
        cnt += 1
```

```python
# комментарии для каких-то признаков признака

cnt = 1
print('Комментарии к некоторым признакам:')
for key in hdr1.keys():
    if key[:2] == 'TC':
        print('    ', key[5:], '-', hdr1[key]) 
        cnt += 1
```

```python
# работа с данными

import numpy as np
import pandas as pd

data = hdul[1].data

# получить numpy массив первого столбца
data.field(0) # можно data.field('ra')
display(data.field(0))

# конвертация в датафрем
RCSED_V2 = pd.DataFrame(np.array(data).byteswap().newbyteorder())
display(RCSED_V2.head())
```

```python
RCSED_V2.describe()
```

```python
data = hdul[0].data
```

```python
import matplotlib.pyplot as plt

plt.hist(data, bins=30)
plt.grid()
plt.show()
```

```python
hdul.close()
```
