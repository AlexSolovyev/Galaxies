# VAE

```python
class VAE(nn.Module):
    def __init__(self, x_dim, h_dim1, h_dim2, z_dim):
        super().__init__()
        
        self.first_block = nn.Sequential(
            nn.Linear(x_dim, h_dim1),
            nn.LeakyReLU(0.2, inplace = True),
            nn.BatchNorm1d(h_dim1),
            nn.Linear(h_dim1, h_dim2),
            nn.LeakyReLU(0.2, inplace = True),
            nn.BatchNorm1d(h_dim2),
        )

        self.fc31 = nn.Linear(h_dim2, z_dim)
        self.fc32 = nn.Linear(h_dim2, z_dim)

        self.fc4 = nn.Sequential(
            nn.Linear(z_dim, h_dim2),
            nn.LeakyReLU(0.2, inplace = True),
            nn.BatchNorm1d(h_dim2),
        )
        self.fc5 = nn.Sequential(
            nn.Linear(h_dim2, h_dim1),
            nn.LeakyReLU(0.2, inplace = True),
            nn.BatchNorm1d(h_dim1),
        )
        self.fc6 = nn.Sequential(
            nn.Linear(h_dim1, x_dim),
        )
        
    def encoder(self, x):
        h = self.first_block(x)
        return self.fc31(h), self.fc32(h)
    
    def sampling(self, mu, log_var):
        std = torch.exp(0.5*log_var)
        eps = torch.randn_like(std)
        return eps.mul(std).add_(mu)
        
    def decoder(self, z):
        h = self.fc4(z)
        h = self.fc5(h)
        return torch.sigmoid(self.fc6(h))
    
    def forward(self, x):
        mu, log_var = self.encoder(x)
        z = self.sampling(mu, log_var)
        return self.decoder(z), mu, log_var
```

```python
device = torch.device("cuda:1" if torch.cuda.is_available() else "cpu")

vae = VAE(x_dim=x.shape[1], h_dim1=128, h_dim2=128, z_dim=2)
vae.to(device)

optimizer = optim.Adam(vae.parameters())
train_loader = torch.utils.data.DataLoader(
    dataset=x,
    batch_size=2**10,
    shuffle=True,
    num_workers=20,
)

def loss_function(recon_x, x, mu, log_var):
    #criterion = nn.MSELoss().cuda()
    #L1 = criterion(recon_x, x.view(-1, x.shape[1]))
    KL = F.kl_div(recon_x, x)
    KLD = -0.5 * torch.sum(1 + log_var - mu.pow(2) - log_var.exp())
    return KL + KLD

def train(epoch):
    vae.train()
    train_loss, cnt = 0, 0
    for i, batch in enumerate(train_loader):
        cnt += 1

        data = batch.requires_grad_(True).to(device)

        optimizer.zero_grad()

        recon_batch, mu, log_var = vae(data)

        loss = loss_function(recon_batch, data, mu, log_var)

        loss.backward()

        train_loss += loss.item()

        optimizer.step()
        
        if cnt % 100 == 1:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, cnt * len(data), len(train_loader.dataset),
                100. * cnt / len(train_loader), loss.item() / len(data)))
    print('====> Epoch: {} Average loss: {:.4f}'.format(epoch, train_loss / len(train_loader.dataset)))
```

```python
for epoch in tqdm(range(7)):
    train(epoch)
```

# GAN

```python
class Generator(nn.Module):
    def __init__(self, x_dim, h_dim1, h_dim2, z_dim):
        super().__init__()
        
        self.enc = nn.Sequential(
            nn.Linear(x_dim, h_dim1),
            nn.LeakyReLU(0.2, inplace = True),
            nn.BatchNorm1d(h_dim1),
            nn.Linear(h_dim1, h_dim1),
            nn.LeakyReLU(0.2, inplace = True),
            nn.BatchNorm1d(h_dim1),
            nn.Linear(h_dim1, z_dim),
            nn.BatchNorm1d(z_dim),
        )

        self.dec = nn.Sequential(
            nn.Linear(z_dim, h_dim2),
            nn.LeakyReLU(0.2, inplace = True),
            nn.BatchNorm1d(h_dim2),
            nn.Linear(h_dim2, h_dim2),
            nn.LeakyReLU(0.2, inplace = True),
            nn.BatchNorm1d(h_dim2),
            nn.Linear(h_dim2, x_dim),
        )
    
    def forward(self, x):
        z = self.enc(x)
        return torch.sigmoid(self.dec(z)), z
    
class Discriminator(nn.Module):
    def __init__(self, x_dim, h_dim):
        super().__init__()
        
        self.fcn = nn.Sequential(
            nn.Linear(x_dim, h_dim),
            nn.LeakyReLU(0.2, inplace = True),
            nn.Linear(h_dim, 1),
        )

    def forward(self, x):
        out = self.fcn(x)
        return torch.sigmoid(out)
```

```python
device = torch.device("cpu")

g = Generator(x_dim=x.shape[1], h_dim1=128, h_dim2=128, z_dim=2)
g.to(device)

d = Discriminator(x_dim=x.shape[1], h_dim=128)
d.to(device)

g_optimizer = optim.Adam(g.parameters())
d_optimizer = optim.Adam(d.parameters())

bce_loss = nn.BCELoss().to(device)
l1_loss = nn.L1Loss().to(device)

train_loader = torch.utils.data.DataLoader(
    dataset=x,
    batch_size=2**16,
    shuffle=True,
    num_workers=16,
)

g.train()
d.train()
def train(epoch):    
    train_g_loss, train_d_loss, cnt = 0, 0, 0
    for i, batch in enumerate(train_loader):
        cnt += 1
        
        x = batch.requires_grad_(True).to(device)

        # шаг генератора
        g_optimizer.zero_grad()

        x_gen, z = g(x)   
        d_fake = d(x_gen)

        loss = bce_loss(d_fake, torch.ones(d_fake.size()).to(device)) + l1_loss(x_gen, x)
        loss.backward()
        train_g_loss += loss.item()

        g_optimizer.step()
        
        if cnt % 100 == 1:
            print('generator loss: {:.10f}'.format(loss.item() / len(x)))

        # шаг дискриминатора
        d_optimizer.zero_grad()

        d_real = d(x)
        r_loss = bce_loss(d_real, torch.ones(d_real.size()).to(device))

        x_gen, z = g(x)
        d_fake = d(x_gen.detach())
        f_loss = bce_loss(d_fake, torch.zeros(d_fake.size()).to(device))

        loss = 0.5 * (r_loss + f_loss)        
        loss.backward()
        train_d_loss += loss.item()
        
        d_optimizer.step()
        
        if cnt % 100 == 1:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tdiscriminator loss: {:.10f}'.format(
                epoch, cnt * len(x), len(train_loader.dataset),
                100. * cnt / len(train_loader), loss.item() / len(x)))
    print('====> Epoch: {} Average g loss: {:.10f}'.format(epoch, train_g_loss / len(train_loader.dataset)))
    print('====> Epoch: {} Average d loss: {:.10f}'.format(epoch, train_d_loss / len(train_loader.dataset)))
```

```python
for epoch in tqdm(range(10)):
    train(epoch)
```

# новый лосс на выход энкодера

```python
device = torch.device("cpu")

g = Generator(x_dim=x.shape[1], h_dim1=128, h_dim2=128, z_dim=2)
g.to(device)

d = Discriminator(x_dim=x.shape[1], h_dim=128)
d.to(device)

g_optimizer = optim.Adam(g.parameters())
d_optimizer = optim.Adam(d.parameters())

train_loader = torch.utils.data.DataLoader(
    dataset=x,
    batch_size=2**16,
    shuffle=True,
    num_workers=16,
)

def train(epoch):
    g.train()
    d.train()
    train_g_loss, train_d_loss, cnt = 0, 0, 0
    for i, batch in enumerate(train_loader):
        cnt += 1
        
        x = batch.requires_grad_(True).to(device)

        # шаг генератора
        g_optimizer.zero_grad()

        x_gen, z = g(x)
        d_fake = d(x_gen)
        
        sim_mtx = F.softmax(z @ z.T)
        mask = torch.diag(torch.ones_like(sim_mtx.diag()))
        sim_mtx = -mask*sim_mtx.diag() + (1. - mask)*sim_mtx

        loss = bce_loss(d_fake, torch.ones(d_fake.size()).to(device)) + l1_loss(x_gen, x) + sim_mtx.mean()
        loss.backward()
        train_g_loss += loss.item()

        g_optimizer.step()
        
        if cnt % 100 == 1:
            print('generator loss: {:.10f}'.format(loss.item() / len(x)))

        # шаг дискриминатора
        d_optimizer.zero_grad()

        d_real = d(x)
        r_loss = bce_loss(d_real, torch.ones(d_real.size()).to(device))

        x_gen, _ = g(x)
        d_fake = d(x_gen.detach())
        f_loss = bce_loss(d_fake, torch.zeros(d_fake.size()).to(device))

        loss = 0.5 * (r_loss + f_loss)        
        loss.backward()
        train_d_loss += loss.item()
        
        d_optimizer.step()
        
        if cnt % 100 == 1:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tdiscriminator loss: {:.10f}'.format(
                epoch, cnt * len(x), len(train_loader.dataset),
                100. * cnt / len(train_loader), loss.item() / len(x)))
    print('====> Epoch: {} Average g loss: {:.10f}'.format(epoch, train_g_loss / len(train_loader.dataset)))
    print('====> Epoch: {} Average d loss: {:.10f}'.format(epoch, train_d_loss / len(train_loader.dataset)))
```

```python
for epoch in tqdm(range(10)):
    train(epoch)
```
